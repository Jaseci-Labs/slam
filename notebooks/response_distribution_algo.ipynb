```python
import uuid
responses = {}
models_responses = []
for model_name, model_data in data.items():
    model_responses = []
    for response in model_data:
        uid = str(uuid.uuid4())
        responses[uid] = response
        model_responses.append(uid)
    models_responses.append({"model_name": model_name, "responses": model_responses})
```

```python
#number of workers
num_workers = 100
#number of questions per worker
number_of_questions_per_worker = 10
# number of models
num_models = 11
# number of responses per model
num_responses = 10

total_num_of_unique_model_pairs = num_models * (num_models - 1) / 2
max_num_of_unique_response_pairs_per_model_pair = num_responses ** 2

print("Number of Unique Model Pairs : ", total_num_of_unique_model_pairs)
print("Maximum Number of Unique Response Pairs per Model : ", max_num_of_unique_response_pairs_per_model_pair)

k = (num_workers * number_of_questions_per_worker / total_num_of_unique_model_pairs)
print("Number of Question Pairs per Model Pair needed : ", k)
print("Number of Responses per Model needed : ", k**0.5)
```

```python
import math
k = math.ceil(k**0.5) ** 2
print("Number of Question Pairs per Model Pair needed : ", k)
print("Number of Responses per Model needed : ", k**0.5)
```

```python
print("Total Number of Questions : ", total_num_of_unique_model_pairs * k)
print("Number of Questions that will not be used : ", total_num_of_unique_model_pairs * k - num_workers * number_of_questions_per_worker)
```

In this case we can go with 3 directions.
1. We can increase the number of workers 130. such that only 75 questions will be not used.
2. Increase the number of questions per worker from 10 to 14 such that all the questions will be used.
3. Just use the 10 questions and ignore the rest of the questions.

As the no. of workers and no. of questions per worker are given (depends on the budget), we can go with the 3rd option.

```python
import itertools
unique_model_pairs = list(itertools.combinations(range(num_models), 2))
print(unique_model_pairs)
```

```python
import random
import math

x = int(math.ceil(num_workers * number_of_questions_per_worker / total_num_of_unique_model_pairs))

question_pairs = []
for model_pair in unique_model_pairs:
    model_a, model_b = model_pair
    model_a_responses = models_responses[model_a]["responses"]
    model_b_responses = models_responses[model_b]["responses"]
    unique_questions_pairs = list(itertools.product(model_a_responses, model_b_responses))
    unique_questions_pair = random.sample(unique_questions_pairs, min(x, len(unique_questions_pairs)))
    for question_pair in unique_questions_pair:
        question_pairs.append((model_pair, question_pair))
```

```python
workers = ["worker" + str(i) for i in range(num_workers)]
```

```python
random.shuffle(question_pairs)
```

```python
distributed_questions = {worker_name: [] for worker_name in workers}
for i in range(num_workers):
    distributed_questions[workers[i]] = question_pairs[i * number_of_questions_per_worker: (i + 1) * number_of_questions_per_worker]
```

```python
distributed_questions
```

```python
# count of model pairs from distributed questions
from collections import Counter

model_pairs = []
question_pairs = []
for questions_of_worker in distributed_questions.values():
    for question_pair in questions_of_worker:
        model_pairs.append(question_pair[0])
        question_pairs.append(question_pair[1])
```

```python
model_pairs_counter = Counter(model_pairs)
model_pairs_counter, len(model_pairs_counter)
```

```python
question_pairs_counter = Counter(question_pairs)
question_pairs_counter, len(question_pairs_counter)
```

```python
# random_approach
questions_done_random = []
for worker in workers:
    for x in range(10):
        random.shuffle(unique_model_pairs)
        random_model_pair = random.choice(unique_model_pairs)
        model_a, model_b = random_model_pair
        model_a_responses = models_responses[model_a]["responses"]
        model_b_responses = models_responses[model_b]["responses"]
        random_question_pair = random.choice(list(itertools.product(model_a_responses, model_b_responses)))
        questions_done_random.append((worker, random_model_pair, random_question_pair))
```

```python
# count of model pairs from distributed questions
from collections import Counter

model_pairs_random = []
question_pairs_random = []
for question_done in questions_done_random:
    model_pairs_random.append(question_done[1])
    question_pairs_random.append(question_done[2])

```
